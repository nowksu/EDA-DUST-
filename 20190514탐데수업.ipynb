```python
import pandas as pd

mise = pd.read_csv('ds_metro.csv', encoding = 'ms949')
mise


ds = mise['측정일시'].astype(str)
mise['날짜'] = ds.str.slice(start = 0, stop = 8).astype(str)
mise['날짜2'] = pd.to_datetime(mise.날짜)

mise['연도2'] = mise['날짜2'].dt.year
mise['월2'] = mise['날짜2'].dt.month
mise['일2'] = mise['날짜2'].dt.day
mise['요일2'] = mise['날짜2'].dt.weekday


import numpy as np
mise['주말여부'] = np.where(mise['요일2']>=5,'주말','평일')

mise0 = mise[(mise['시간'] == 11)] 

mise_fin = mise0.groupby(['시도', '측정일시']).max()

mise_fin.reset_index(inplace = True)

#80->1 150->2     

mise_fin.loc[mise_fin['PM10']<=80,'PM10그룹']=0
mise_fin.loc[mise_fin['PM10']>80,'PM10그룹']=1
mise_fin.loc[mise_fin['PM10']>150,'PM10그룹']=2


mise_fin.loc[mise_fin['PM25']<=35,'PM25그룹']=0
mise_fin.loc[mise_fin['PM25']>35,'PM25그룹']=1
mise_fin.loc[mise_fin['PM25']>75,'PM25그룹']=2


mise_fin['PM10그룹'].value_counts()

mise_fin['PM25그룹'].value_counts()

                                                    
pd.crosstab([mise_fin['시도'],mise_fin['연도2']],[mise_fin['PM10그룹']],normalize = 'index')

                                                    
pd.crosstab([mise_fin['시도'],mise_fin['연도2']],[mise_fin['PM25그룹']],normalize = 'index')

mise_fin10 = pd.crosstab([mise_fin['시도'],mise_fin['연도2'],mise_fin['주말여부']],[mise_fin['PM10그룹']],normalize = 'index')

mise_fin.head(3)

                                                    
mise_fin25 = pd.crosstab([mise_fin['시도'],mise_fin['연도2'],mise_fin['주말여부']],[mise_fin['PM25그룹']],normalize = 'index')

mise_fin10.to_csv('mise_fin10.csv', encoding = 'ms949')
mise_fin25.to_csv('mise_fin25.csv', encoding = 'ms949')

mise_fin2=mise_fin[['시도','연도2','PM10','PM25','SO2','CO','O3','NO2']]

mise_corr =mise_fin2.groupby(['시도','연도2']).corr()

mise_corr

mise_corr.to_csv('mise_corr.csv', encoding = 'ms949')
```
